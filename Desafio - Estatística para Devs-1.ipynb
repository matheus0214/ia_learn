## Estatística para Devs

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
df_faturamento
```

```python
df_faturamento.info()
```

```python
# Ordenar pela data de referência
df_faturamento = df_faturamento.sort_values('data_ref')
```

### Média das vendas

```python
media = df_faturamento['valor'].mean()
print(f'Média das vendas: R$ {media:,.2f}')
```

## Gráfico de Barras (Faturamento por mês)

```python
fig, ax = plt.subplots(figsize=(10, 5))

meses = df_faturamento['data_ref'].dt.strftime('%m/%Y')

bars = ax.bar(meses, df_faturamento['valor'])
ax.set_title('Faturamento por Mês', fontweight='bold')
ax.set_xlabel('Mês')
ax.set_ylabel('Valor (R$)')

ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'R$ {x/1000:.0f}k'))
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## Gráfico de Linha (Faturamento por mês)

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_faturamento['data_ref'], df_faturamento['valor'], marker='o', linewidth=2, markersize=8, color='steelblue')

ax.set_title('Faturamento por Mês', fontsize=14, fontweight='bold')
ax.set_xlabel('Mês')
ax.set_ylabel('Valor (R$)')

ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'R$ {x/1000:.0f}k'))

ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
