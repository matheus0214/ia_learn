```python
import pandas as pd
```

```python
pd.__version__
```

```python
dict_medidas = {
    'idade': [15, 18, 25, 25, 40, 55, 58, 60, 80],
    'altura': [160, 162, 165, 168, 172, 174, 174, 174, 176]
}
```

```python
dict_medidas
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
```

```python
df_medidas
```

## Medidas Resumo

### Medidas de Posição

```python
# média
df_medidas['idade'].mean()
```

```python
# mediana
df_medidas['idade'].median()
```

```python
# moda
df_medidas['idade'].mode()
```

```python
# moda altura
df_medidas['altura'].mode()
```

### Medidas de Disperção

```python
# variãncia
df_medidas['idade'].var()
```

```python
# desvio padrão
df_medidas['idade'].std()
```

```python
# coeficiente de variação
df_medidas['idade'].std() / df_medidas['idade'].mean() * 100
```

```python
df_medidas['altura'].std() / df_medidas['altura'].mean() * 100
```

### Medidas de Forma

```python
# assimetria - idade
df_medidas['idade'].skew()
```

```python
# assimetria - altura
df_medidas['altura'].skew()
```

```python
# curtose - idade
df_medidas['idade'].kurtosis()
```

```python
# curtose - altura
df_medidas['altura'].kurtosis()
```

```python
# obter medidas estatísticas
df_medidas['idade'].describe()
```

### Coeficiente de Correlação

```python
# correlação do dataframe, segundo Pearson
df_medidas.corr(method='pearson')
```

```python
# correlação entre duas variáveis
df_medidas['idade'].corr(df_medidas['altura'])
```

```python
# correlação dataframe, segundo Pearson
df_medidas.corr(method='pearson')
```

### Criar Dataframe de Vendas

```python
dict_vendas = {
    'categoria': ['masculino', 'feminino', 'infantil', 'casa'],
    'valor': [400000, 600000, 250000, 580000],
    'quantidade': [3000, 5000, 1500, 2500]
}
```

```python
df_vendas = pd.DataFrame.from_dict(dict_vendas)
```

```python
df_vendas
```

### Gráfico de Barras

```python
# Vertical
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
# Vertical - Valor (ordernado)
df_vendas.sort_values('valor', ascending=False).plot.bar(x='categoria', y='valor')
```

```python
# Horizontal - Quantidade
df_vendas.plot.barh(x='categoria', y='quantidade')
```

```python
# Vertical - Quantidade ordenado (Ranking)
df_vendas.sort_values('quantidade').plot.barh(x='categoria', y='quantidade')
```

### Gráfico de Dispersão

```python
df_medidas.plot.scatter(x='idade', y='altura')
```

### Gráfico Boxplot (Diagrama de Caixa)

```python
df_medidas['idade'].plot.box()
```

```python
df_medidas['altura'].plot.box()
```

### Gerar Dataframe dados temporais

```python
dict_faturamento = {
    'data_ref': ['2019-01-01', '2020-01-01', '2021-01-01', '2022-01-01', '2023-01-01'],
    'valor': [400000, 800000, 500000, 800000, 900000]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
df_faturamento.info()
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
df_faturamento.info()
```

### Gráfico de Linhas

```python
df_faturamento.plot.line(x='data_ref', y='valor')
```
