## Unificar DataFrames de Customers, Services e Contracts

```python
# Contar quantidade de registros por DataFrame
len(customers)
```

```python
len(df_services)
```

```python
len(df_contracts)
```

```python
# Renomear colunas
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
# Unificar DataFrame de Customers com Services, criando um terceiro DataFrame
df_temp = customers.merge(df_services, on='IDCliente', how='inner', validate='one_to_one')
```

```python
df_temp.info()
```

```python
# Unificar o df_temp com Contracts, usando colunas de junção com nomes distintos
df_churn_temp = df_temp.merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'], how='inner', validate='one_to_one')
```

```python
df_churn_temp.info()
```

```python
# Unificar os três DataFrames ao mesmo tempo, com colunas com nomes distintos
df_churn = customers.merge(
    df_services,
    on=['IDCliente'],
    how='inner',
    validate='one_to_one'
).merge(
    df_contracts,
    left_on=['IDCliente'],
    right_on=['customerID'],
    how='inner',
    validate='one_to_one'
)
```

```python
df_churn.info()
```

```python
# Removendo colunas de um DataFrame
df_churn = df_churn.drop(columns=['customerID'])
```

```python
df_churn.info()
```

## Imputação de Valores ausentes

```python
# Pegar os indices que contem valores ausentes
index_with_missing_values = df_churn[df_churn.isna().any(axis=1)].index
```

```python
# Preencher todos os valores ausentes com 0
df_churn.fillna(0, axis=1).iloc[index_with_missing_values]
```

```python
# Preencher valores padrão conforme a coluna
df_churn.fillna(value={'TotalCharges': 0, 'Genero': 'Não declarado'})
```
