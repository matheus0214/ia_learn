## Transformação de Dados

```python
# Transformar colunar TotalCharges de String para Float - abordagem 1 com astype

df_contracts['TotalCharges'].astype(float)
```

```python
# Transformar colunar TotalCharges de String para Float - abordagem 2 com to_numeric

pd.to_numeric(df_contracts['TotalCharges'])
```

```python
# Transformar colunar TotalCharges de String para Float - abordagem 3 com to_numeric com coerce

df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'], errors='coerce')
```

```python
# Mostrar estrutura/schema do DataFrame
df_contracts.info()
```
